```python
data=soup.find_all("div",class_="blog--single__content column--3-4 u-spacing-third")
data
```

```python
mon_n=[]
for i in data:
    for j in i.find_all("p"):
        k=j.text.replace("\n","")
        mon_n.append(k)
```

```python
mon_name=[]

for i in range(1,30,3):
    mon_name.append(mon_n[i])
mon_name
```

```python
desc=[]
for i in range(2,30,3):
    desc.append(mon_n[i])
desc
```

```python
img_url= []
for item in soup.find_all("img"):
    img_url.append(item["src"])
img_url
```

```python
image=[]
for i in range(6,26,2):
    image.append(img_url[i])
image
```

```python
df=pd.DataFrame(list(zip(mon_name,desc,image)),columns=["Monument Name","Description","Image"])
df
```
