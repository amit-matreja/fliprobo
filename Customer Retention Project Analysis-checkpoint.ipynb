# Analysing Positive and Negative Reviews to find some insights

```python
pos=df1.drop(['Longer time to get logged in (promotion, sales period)',
       'Longer time in displaying graphics and photos (promotion, sales period)',
       'Late declaration of price (promotion, sales period)',
       'Longer page loading time (promotion, sales period)',
       'Limited mode of payment on most products (promotion, sales period)',
       'Longer delivery period', 'Change in website/Application design',
       'Frequent disruption when moving from one page to another'],axis=1)
pos.columns
```

```python
# Lets count positive and Negetive reviews of websites/online stores/
website=['Amazon.in','Flipkart.com','Paytm.com','Myntra.com','Snapdeal.com']
#Creating a function that can give the count of positive and negative review
def pos_neg(x,y):
  c1=[]
  w_site=[]
  count1=[]
  for i in x.columns:                     #x represent the original columns
    for j in y:                           #y represent the site only
      count=len(x[x[i].str.contains(j)])
      c1.append(i)
      w_site.append(j)
      count1.append(count)
  positive=pd.DataFrame([c1,w_site,count1])  #Converting all the values in column
  return positive
```

```python
positive1=pos_neg(pos,website)
positive1
```

```python
positive1=positive1.T
p1=positive1.rename(columns={0:'Positive Reviews',1:'Website',2:'count'})
p1
```

* Here in the table we can see all the positive reviews with total no. of count. we can clearly see that Amazon.in has got
maximum reviews for easy to use website and application followed by flipkart.com and so on.

```python
# Lets check the negative reviews as well.
neg=df[['Longer time to get logged in (promotion, sales period)',
       'Longer time in displaying graphics and photos (promotion, sales period)',
       'Late declaration of price (promotion, sales period)',
       'Longer page loading time (promotion, sales period)',
       'Limited mode of payment on most products (promotion, sales period)',
       'Longer delivery period', 'Change in website/Application design',
       'Frequent disruption when moving from one page to another']]
```

```python
negative=pos_neg(neg,website)
```

```python
negative=negative.T
n1=negative.rename(columns={0:'Negative reviews',1:'Website',2:'Count'})
n1
```

* Here we can see that Amazon is on top for negative reviews as well for longer time to get logged in.

```python
# lets visualize positive and negative reviews using the barplot.
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
sns.barplot(x='Website',y='count',data=p1)
plt.title("Positive Count",fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.subplot(1,2,2)
sns.barplot(x='Website',y='Count',data=n1)
plt.title("Negative Count",fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.show()
```

* Here Using the bar graph we can clearly see that Amazon has got maximum no. of Positive reviews as well as maximum no. of negative reviews but count of negative reviews is far less than the count of positive reviews, same for flipkart who stands 2nd for both positive and negative reviews, same goes for myntra as well, but paytm and snapdeal are two online stores which have more of negative reviews as compared to positive reviews.
